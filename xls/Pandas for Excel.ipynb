```python
import pandas as pd;
import numpy as np;
```

```python
xls = pd.read_excel('https://github.com/datagy/pivot_table_pandas/raw/master/sample_pivot.xlsx')
```

```python
xls.head(20)
```

```python
#Note the difference
# The indexing operator with an array returns a DataFrame
# While a single column is a Series

type(xls['Sales']) # pandas.core.series.Series

type(xls[['Sales']]) # pandas.core.frame.DataFrame
```

```python
xls.sort_values(by='Sales', ascending=False)
```

```python
# Sort by multiple columns
xls.sort_values(by=['Sales', 'Units'], ascending=False)
```

```python
# Filter
# To filter need to pass an array of booleans. Each row which is True is returned
# xls['Region'] == 'North' returns an array of boolean

xls[xls['Region'] == 'North']
```

```python
# Multiple statments combined using braces (). Take note of & and | operators
xls[(xls['Region'] == 'North') & (xls['Sales'] > 200)]
```

```python
xls[(xls['Region'] == 'North') | (xls['Sales'] > 900)]
```

```python
#Reorder

xls[['Region', 'Type', 'Sales', 'Units']]
```

```python
# Pivot table

pd.pivot_table(xls, index=['Region'] , columns = ['Type'] , values = ['Units', 'Sales'], aggfunc={'Sales': np.sum, 'Units': np.mean})
```
